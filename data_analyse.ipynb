```python
def read_data(path):
    """
    read_data处理

    path:  数据所在的文件路径

    return: 返回值是处理后的结果
    """
    f = hdf5storage.loadmat(path)

    if path == '../data/All_MM_MA_0.mat':
        element_name = 'MM_average_MA_0'
    elif path == '../data/All_MM_MA_04.mat':
        element_name = 'MM_average_MA_04'

    src_data = np.array(f[element_name])

    return src_data
```

```python
path = ['../data/All_MM_MA_0.mat', '../data/All_MM_MA_04.mat']

a = read_data(path[0])

b = a[1]

c = a[2]

d = [b,c]

# mean = np.mean(b)
# variance = np.var(b)

# 绘制箱线图
sns.boxplot(data=d, showfliers=False)

# 添加标题

plt.title('Data')

plt.show()
```

```python
import matplotlib.pyplot as plt

# 数据
b = [1, 2, 3, 4, 5]
c = [6, 7, 8, 9, 10]

data = [b, c]  # 将数据放入列表中

# 箱线图
fig, ax = plt.subplots()

boxprops = dict(color='lightblue', linewidth=2)  # 箱体属性
whiskerprops = dict(color='blue', linewidth=2)  # 触须属性
flierprops = dict(marker='o', markerfacecolor='red', markersize=8, markeredgecolor='black')  # 异常值属性

ax.boxplot(data, boxprops=boxprops, whiskerprops=whiskerprops, flierprops=flierprops)

# 标题和标签
ax.set_title('Customized Box Plot')
ax.set_xticklabels(['b', 'c'])
ax.set_ylabel('Values')

# 显示图形
plt.show()
```
